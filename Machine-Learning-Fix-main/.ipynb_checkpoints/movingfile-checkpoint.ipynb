```python
import os
import shutil
```

```python
path = r"C:\Users\chris\OneDrive\Documents\GitHub\Machine-Learning-Fix\fases\\"
newpath=r"C:\Users\chris\OneDrive\Documents\GitHub\Machine-Learning-Fix\fasespart\\"
SalmoDir = os.makedirs(os.path.join(newpath,"salmo"))
HealthyDir = os.makedirs(os.path.join(newpath,"healthy"))
CocciDir = os.makedirs(os.path.join(newpath,"cocci"))
NcdDir = os.makedirs(os.path.join(newpath,"ncd"))
```

```python
os.listdir(path)
```

```python
for file in os.listdir(path):
    file_split = file.split('.')
    filename = file_split[0]
    try:
        if filename in "pcrsalmo":
            filename = file_split[0]+file_split[1]+file_split[2]
            shutil.move(path+filename,SalmoDir+filename)
        elif filename in "pcrhealthy":
            filename = file_split[0]+file_split[1]+file_split[2]
            shutil.move(path+filename,HealthyDir+filename)
        elif filename in "pcrcocci":
            filename = file_split[0]+file_split[1]+file_split[2]
            shutil.move(path+filename,CocciDir+filename)
        elif filename in "pcrncd":
            filename = file_split[0]+file_split[1]+file_split[2]
            shutil.move(path+filename,NcdDir+filename)
        else:
            pass
    except Exception:
        pass
```
